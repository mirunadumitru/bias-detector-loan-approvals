```python
import json
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, confusion_matrix, classification_report,
    RocCurveDisplay, PrecisionRecallDisplay
)

RANDOM_STATE = 42
```

```python
df = pd.read_csv("../data/processed/loan_approval_clean.csv")

df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

df['loan_status'] = df['loan_status'].astype(str).str.strip().str.lower()
valid = {"approved":1, "rejected":0}
df = df[df['loan_status'].isin(valid)].copy()
df['loan_status_bin'] = df['loan_status'].map(valid)

drop_cols = ['loan_status', 'loan_status_bin', 'loan_id']
X = df.drop(columns=[c for c in drop_cols if c in df.columns], errors='ignore')
y = df['loan_status_bin']

categorical = [c for c in ['education','self_employed'] if c in X.columns]
numeric = [c for c in [
    'no_of_dependents','income_annum','loan_amount','loan_term','cibil_score',
    'residential_assets_value','commercial_assets_value','luxury_assets_value','bank_asset_value'
] if c in X.columns]

print("X shape:", X.shape, "y shape:", y.shape)
print("Categorical:", categorical)
print("Numeric:", numeric)
print("Class balance:", y.value_counts(normalize=True).round(3))
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=RANDOM_STATE, stratify=y
)
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), numeric),
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical),
    ],
    remainder='drop'
)
```

```python
logreg = Pipeline([
    ("prep", preprocessor),
    ("clf", LogisticRegression(max_iter=2000, class_weight="balanced", random_state=RANDOM_STATE))
])

rf = Pipeline([
    ("prep", preprocessor),
    ("clf", RandomForestClassifier(
        n_estimators=400, max_depth=None, n_jobs=-1,
        class_weight="balanced_subsample", random_state=RANDOM_STATE))
])
models = {"logreg": logreg, "random_forest": rf}
```

```python
def eval_model(name, model, X_tr, y_tr, X_te, y_te):
    model.fit(X_tr, y_tr)
    y_pred = model.predict(X_te)
    y_prob = getattr(model, "predict_proba", None)
    proba = y_prob(X_te)[:,1] if y_prob else None

    metrics = {
        "accuracy": accuracy_score(y_te, y_pred),
        "precision": precision_score(y_te, y_pred),
        "recall": recall_score(y_te, y_pred),
        "f1": f1_score(y_te, y_pred),
    }
    if proba is not None:
        metrics["roc_auc"] = roc_auc_score(y_te, proba)

    print(f"\n=== {name} ===")
    print(classification_report(y_te, y_pred, digits=3))
    print("Confusion matrix:\n", confusion_matrix(y_te, y_pred))
    if proba is not None:
        RocCurveDisplay.from_predictions(y_te, proba); plt.title(f"ROC — {name}"); plt.show()
        PrecisionRecallDisplay.from_predictions(y_te, proba); plt.title(f"PR — {name}"); plt.show()
    return metrics, model
```

```python
all_metrics = {}
fitted = {}

for name, mdl in models.items():
    m, f = eval_model(name, mdl, X_train, y_train, X_test, y_test)
    all_metrics[name] = {k: float(v) for k, v in m.items()}
    fitted[name] = f

print("\nSummary metrics:", json.dumps(all_metrics, indent=2))
```

```python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
for name, mdl in models.items():
    scores = cross_val_score(mdl, X, y, scoring="f1", cv=cv, n_jobs=-1)
    print(f"{name} 5-fold F1: mean={scores.mean():.3f} ± {scores.std():.3f}")
```

```python
def score_key(m):
    return (m.get("roc_auc", 0.0), m["f1"])

best_name = max(all_metrics, key=lambda k: score_key(all_metrics[k]))
best_model = fitted[best_name]
print("Best model:", best_name, "metrics:", all_metrics[best_name])

import os
os.makedirs("../models", exist_ok=True)
os.makedirs("../reports", exist_ok=True)

joblib.dump(best_model, f"../models/{best_name}.pkl")
with open("../reports/metrics.json", "w") as f:
    json.dump(all_metrics, f, indent=2)

with open("../models/feature_config.json", "w") as f:
    json.dump({"categorical": categorical, "numeric": numeric}, f, indent=2)
```

```python
from sklearn.model_selection import GridSearchCV

grid = {
    "clf__C": [0.1, 1.0, 3.0, 10.0],
    "clf__penalty": ["l2"],
    "clf__solver": ["lbfgs"],
}
gs = GridSearchCV(
    estimator=logreg, param_grid=grid, scoring="f1",
    cv=5, n_jobs=-1, verbose=1
)
gs.fit(X_train, y_train)
print("Best params:", gs.best_params_, "Best F1:", gs.best_score_)
best_lr = gs.best_estimator_
_ = eval_model("logreg_tuned", best_lr, X_train, y_train, X_test, y_test)
```

```python
from sklearn.metrics import precision_recall_curve

proba = best_model.predict_proba(X_test)[:,1]
prec, rec, thr = precision_recall_curve(y_test, proba)
f1s = 2*prec*rec/(prec+rec + 1e-9)
best_idx = np.nanargmax(f1s)
best_thr = thr[best_idx] if best_idx < len(thr) else 0.5
print(f"Best F1 threshold ≈ {best_thr:.3f}, F1={f1s[best_idx]:.3f}")

with open("../models/threshold.txt","w") as f:
    f.write(str(best_thr))
```
